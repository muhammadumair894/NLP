```python
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
sentence="studies studying cries going goes gone playing plays"
words = word_tokenize(sentence)
ps = PorterStemmer()
temp = ''
for w in words:
  rootWord=ps.stem(w)
  #print(rootWord)
  temp = temp + rootWord + " "

print(temp)
```

```python
import nltk
from nltk.stem import 	WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
text = "studies studying cries going goes gone playing plays"
#text="Hello Guru99, You have to build a very good site and I love visiting your site."
tokenization = nltk.word_tokenize(text)
for w in tokenization:
  print("Lemma for {} is {}".format(w, wordnet_lemmatizer.lemmatize(w)))
```
